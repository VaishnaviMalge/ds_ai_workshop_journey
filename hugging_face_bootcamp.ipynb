#Text to Image using Diffusion Pipeline

```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "sd-legacy/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "innocent child playing with a ball in gardon. A cute puppy adorably looking towards her."
image = pipe(prompt).images[0]

image.save("child_puppy_ball.png")
```

```python
#import gradio as gr
from diffusers import StableDiffusionPipeline
import torch

model_id = "sd-legacy/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

def generate_image(prompt):
  image = pipe(prompt).images[0]
  image.save(f'{prompt}.png')
  return image

# UI

gr.Interface(
    fn = generate_image,
    inputs = gr.Textbox(label ='Enter your prompt',placeholder = 'e.g.innocent child'),
    outputs = gr.Image(type = 'pil', label ='Generated Image'),
    title = 'Text to Image Generator using Stable Diffusion',
    description = 'Enter your promt to see your imagination gets visualized.'
    ).launch()
```
